```python
import pandas as pd
from IPython.display import display
```

```python
movies = pd.read_csv('../archive/IMDb movies.csv')
rating = pd.read_csv('../archive/IMDb ratings.csv')
movies.head(1)
```

```python
rating.head(1)
```

```python
mov = movies.loc[:,['imdb_title_id','title','year','date_published','genre']]
rate = rating.loc[:,['imdb_title_id', 'total_votes', 'weighted_average_vote']]
mov.head()
rate.head()
```

```python
imdb = mov.merge(how='outer', right=rate, right_on='imdb_title_id', left_on='imdb_title_id')
imdb.head(1)
```

```python
netflix = pd.read_csv('../IMDB movies.csv/netflix_titles.csv')
```

```python
netflix.head(1)
```

```python
net_imdb = imdb.merge(how= 'inner', right = netflix, right_on = ['title','release_year'], left_on = ['title','year'])
```

```python
net_imdb.head(1)
```

# Genre : Action

```python
net_imdb.loc[net_imdb['genre'].str.contains('Action')].sort_values('weighted_average_vote', ascending=False).sort_values('total_votes', ascending=False).head(5)
```

## Best action movies that has total votes higher than 300,000

```python
net_imdb.loc[(net_imdb['total_votes']>300000)&(net_imdb['genre'].str.contains('Action'))].sort_values('weighted_average_vote', ascending=False).head(5)
```
